```python
import fsub_extractor.utils.fury_viz as fv
from fsub_extractor.functions.streamline_scalar import streamline_scalar as ss
import pandas as pd 
import os.path as op
import nibabel as nib
```

We can visualize the outputs of FSuB-Extractor by defining fury actors for the streamlines and fROIs. Here we will create an axial view of the streamlines connecting to mOTS-words in blue and mFus-faces in red. 

```python
sub='subj04'
hem='lh'
rois = ['mOTS_words','mFus_faces']
colors = [[0,0,1],[1,0,0]]

extractor_dir = op.join('..','extractor_outputs',sub)
ref_anat = op.join('..','freesurfer',sub,'mri','T1.nii.gz')
streamline_actors = [None]*len(rois)
roi_actors = [None]*len(rois)

for r in range(len(rois)):
    tck = op.join(extractor_dir,'dwi',hem+'_'+rois[r]+'_extracted.tck')
    streamline_actors[r] = fv.define_streamline_actor(tck,ref_anat,colors[r])
    roi_path = op.join(extractor_dir,'func',hem+'_'+rois[r]+'_'+hem+'_'+rois[r]+'.projected_gmwmi_intersected.nii.gz')
    roi_actors[r]=fv.define_roi_actor(roi_path,colors[r],opacity=1)

slice_actor = [None]*1
slice_actor[0] = fv.define_slice_actor(ref_anat,view='axial',offset=-.15)
fname = op.join('..','figures', sub+'_'+hem+'_mFus_mOTS_AF.png')
fv.visualize_bundles(streamline_actors,0,slice_actor,roi_actors,camera_angle='axial',hemi = hem,filename=fname)
```

We can also visualize the two fROI case, by ploting IFS words and mOTS words and the streamlines that connect between them. 

```python
sub='subj04'
hem='lh'
prefix = 'lh_mOTS_IFS'
rois = ['mOTS_words','IFS_words']
colors = [[0,0,1],[0,0,1]]

extractor_dir = op.join('..','extractor_outputs',sub)
ref_anat = op.join('..','freesurfer',sub,'mri','T1.nii.gz')
streamline_actors = [None]*len(rois)
roi_actors = [None]*len(rois)

for r in range(len(rois)):
    roi_path = op.join(extractor_dir,'func',prefix+'_'+hem+'_'+rois[r]+'.projected_gmwmi_intersected.nii.gz')
    roi_actors[r]=fv.define_roi_actor(roi_path,colors[r],opacity=1)
    
tck = op.join(extractor_dir,'dwi',prefix+'_extracted.tck')
streamline_actors[r] = fv.define_streamline_actor(tck,ref_anat,[0,0,.8])
slice_actor = [None]*1
slice_actor[0] = fv.define_slice_actor(ref_anat,view='saggital')
fname = op.join('..','figures', sub+'_'+hem+'_IFS_mOTS_AF_sagittal.png')
fv.visualize_bundles(streamline_actors,0,slice_actor,roi_actors,camera_angle='saggital',hemi = hem,filename=fname)
```

We might also want to compare scalar values for the FSuB connecting to different fROIs. Here, we will calculate the scalar values for mOTS-words and mFus-faces. 

```python
sub='subj04'
rois=['lh_mOTS_words','lh_mFus_faces']
extractor_dir = op.join('..','extractor_outputs')
data_dir = op.join('..','data',sub)
scalar_paths = op.join('..','data',sub,'dwi','fa.nii.gz')
scalar_names='FA'
n_points = 100
out_dir=extractor_dir
overwrite=0

for roi in rois:
    tract = op.join(extractor_dir,sub,'dwi',roi+'_extracted.tck')
    out_prefix=roi
    ss(
        sub,
        tract,
        scalar_paths,
        scalar_names,
        n_points,
        out_dir,
        out_prefix,
        overwrite,
    )
```
